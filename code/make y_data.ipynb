```python
#원본 댓글 데이터 가져오기 
f2=open('Us C.txt',mode='rt')

#만약, y_data를 처음부터 다시 만들고 싶으면 아래 3줄 코드를 밑 셀에 오려붙이세요. 
step=1
y_index=0
y_data=[]
```

```python
#y_data 입력하기. 중간에 다른 문자를 입력해서 프로그램이 멈췄을 경우, 
#그 전까지 입력한 y_data는 그대로 저장되어 있습니다. 
#q을 눌러도 이 셀을 다시 실행하면, 이어서 진행할 수 있습니다. 
door=True
while door:
    line=f2.readline()
    if not line: break
    while True:
        print(str(step)+') 댓글:'+line)
        
        ans=input("(q/요청:0,부정:1,질문:2,긍정:3) 입력:")
        if(ans=='q'): #quit을 누른 경우 
            door=False
            break 
        elif(int(ans)==0 or int(ans)==1 or int(ans)==2 or int(ans)==3) :
            y_data.append(int(ans))
            ans2=input("다시 입력하시겠습니까?(y/n):")
            if ans2=='y': continue
            elif ans2=='n': 
                step+=1
                y_index+=1
                break
            else :
                print("잘못 입력함. 다시 입력해주세요.")
                continue
        else:      
            print("번호를 잘못 입력함.")
            continue
    if(door==False):
        print("--프로그램 종료--")
        print(y_data)
```
