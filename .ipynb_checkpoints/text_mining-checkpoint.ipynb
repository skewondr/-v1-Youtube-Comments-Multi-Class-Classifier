```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
import nltk  
nltk.download('punkt')
from nltk.tokenize import word_tokenize  
import re

#import nltk ì•ˆë˜ë©´ 
#pip uninstall nltk
#pip install -U nltk
```

```python
#í¬ë¡¤ë§í•œ txtíŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
f1=open('content.txt',mode='rt',encoding='utf-8')

#í‘œê¸°ê°€ ë‹¤ë¥¸ ë‹¨ì–´ë“¤ì˜ í†µí•© ex) US=USA
#ëŒ€ì†Œë¬¸ì í†µí•© : ë¬¸ì¥ ì²« ë‹¨ì–´ë§Œ ì†Œë¬¸ìë¡œ ë°”ê¾¸ê¸° 
#ë¶ˆìš©ì–´, ë“±ì¥ ë¹ˆë„ê°€ ì ì€ ë‹¨ì–´, ì´ëª¨í‹°ì½˜ ì§€ìš°ê¸°

#ë¶ˆìš©ì–´ ì œê±°ë¥¼ ìœ„í•œ ëª¨ë“ˆ
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))

#ì´ëª¨í‹°ì½˜ ì œê±°ë¥¼ ìœ„í•œ ëª¨ë“ˆ
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F937\u200D\u2642\uFE0F"
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
#comment_list = []
#for i in range(len(comment_data)):
#    comment_list.append(comment_data['comment'].iloc[i])

```

```python
#raw_data=[] #ë¬¸ì¥ ì›ë³¸
text_data=[] #í† í°í™” ëœ ë¬¸ì¥


while True:
    line=f1.readline()
    #raw_data.append(line)
    #ëª¨ë“  ë‹¨ì–´ ì†Œë¬¸ìë¡œ ë§Œë“¤ê¸°
    line=line.lower()
    line=re.sub('[-=+,#/\:^$.@*\"â€»~&%ã†!ã€\\â€˜|\(\)\[\]\<\>`â€œ\'â€¦ã€‹;]','',line)
    if not line: break
    #ë¬¸ì¥ í† í°í™” í•˜ê¸° 
    #TreebankWordTokenizer: 1. -ìœ¼ë¡œ êµ¬ì„±ëœ ë‹¨ì–´ëŠ” í•˜ë‚˜ë¡œ 2. doesn't=does+n't
    line=re.sub(emoji_pattern,'',line)
    line2=tokenizer.tokenize(line) 
    text_data.append(line2)
    
#print(raw_data)
print(text_data)
```

```python
# ì–´ê°„ì¶”ì¶œ(stemming) - í’ˆì‚¬ì˜ ì •ë³´ê°€ ë³´ì¡´ë˜ì§€ ì•ŠìŒ.  
# vs í‘œì œì–´ ì¶”ì¶œ(lemmatization) - í’ˆì‚¬ì˜ ì •ë³´ê°€ ë³´ì¡´ë¨.
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 
n=WordNetLemmatizer()

text_data2=[]

for l in text_data:
    result=[]
    for w in l:
        #í‘œì œì–´ ì¶”ì¶œ
        n.lemmatize(w) 
        #ë¶ˆìš©ì–´ ì œê±°
        if w not in stop_words:
            result.append(w)
    text_data2.append(result)

    
print(text_data2)
   
```

```python
import csv

#ë¦¬ìŠ¤íŠ¸ ì €ì¥í•˜ê¸°   
with open('token.csv','w') as writeFile:
    writer = csv.writer(writeFile)
    writer.writerows(text_data2)

writeFile.close()
```
