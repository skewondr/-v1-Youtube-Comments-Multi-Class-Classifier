```python
def showBoard(answer,horse):
    r,c=answer.split(",")
    r=int(r)-1
    c=int(c)-1
    if board[r][c]!='?': 
        print("이미 다른 player가 선택한 위치입니다.")
        return False
    elif horse==1:
        board[r][c]='O'
    elif horse==2:
        board[r][c]='X'
    
    for row in range(3):
        for col in range(3):
            print(board[row][col], end=' ')
        print()
    return True

def isOk(answer):
        row_input,col_input=answer.split(",")
        row_input=int(row_input)
        col_input=int(col_input)
        if row_input>0 and row_input<=3 and col_input>0 and col_input<=3:
            return True
        else :
            print("위치가 유효하지 않습니다. 행렬 인덱스는 1~3까지 유효합니다.")
            return False
```

```python
#Winner가 나오지 않으면 게임을 반복한다. 
def winner(horse):
            
    win=False
    for i in range(3):
        #check row
        if board[i][0]==horse and board[i][1]==horse and board[i][2]==horse: 
            win=True
            break
        #check col
        if board[0][i]==horse and board[1][i]==horse and board[2][i]==horse:
            win=True
            break
    #check diagonal
    if board[0][0]==horse and board[1][1]==horse and board[2][2]==horse:
        win=True
    if board[0][2]==horse and board[1][1]==horse and board[2][0]==horse:
        win=True
    
    if win:
        print('The Winner is Player {}'.format(horse))
        print('게임을 종료합니다...')
        return False
    else :
        for j in range(3):
            for k in range(3):
                if board[j][k]=='?': return True
        print('승자가 없습니다.')
        print('게임을 종료합니다...')
        return False
```

```python

print("----Tic Tac Toe Game----\n\n")

door=True
#3*3행렬을 0으로 초기화 한다. 
board = [['?' for col in range(3)] for row in range(3)]
while door:
    while door:
        where1=input("player 1 입력하세요. (예: 2,3) :") 
        if isOk(where1) and showBoard(where1,1): 
            door=winner('O')
            break;
        else : continue;
    
    while door: 
        where2=input("player 2 입력하세요. (예: 2,3) :")
        if isOk(where2) and showBoard(where2,2):   
            door=winner('X')
            break;
        else : continue;
    
```
